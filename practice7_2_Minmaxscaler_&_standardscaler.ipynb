```
# normalization : scale을 0과 1 사이로 만들어주는 것 의미
# Minmaxscaler : 해당 data를 원하는 범위로 조정하는 것(min과 max를 파악해서 조정하는,,)
  # minmaxscaler에서 feature_range를 0과 1 사이로 만드는 것 == normalization
  # result >> feature의 분포가 0과 1 사이 안으로 들어온다.
# Standardization
  #고등학교 통계과목에서 배운 표준화를 의미
  #해당 data feature의 평균과 표준편차를 구해서 "(feature값 - 평균) / 표준편차 " 연산을 가해준다.
  # result >> feature의 분포가 표준정규분포를 따른다 =N(0,1) (standard normal distribution)
```
